```python
import pandas as pd 
import numpy as np 

import matplotlib.pyplot as plt
plt.style.use("bmh")
%matplotlib inline
```

```python
df = pd.read_csv("lightning_logs/fashion/version_4/relu_epoch.csv")
print(df.shape)
df.head()
```

```python
df_bn_mean = df[[i for i in df.columns if "percent" in i]]
df_bn_mean
```

```python
ncols = 3 
nrows = np.ceil(len(df_bn_mean.columns)/ncols).astype(int)
print(nrows, ncols)
fig, ax = plt.subplots(figsize=(nrows*3, ncols*2), nrows=nrows, ncols=ncols)
cols = df_bn_mean.columns
for i in range(nrows*ncols): 
    if i<len(cols): 
        ax.flat[i].plot(range(len(df_bn_mean)), df_bn_mean[cols[i]].values)
        ax.flat[i].set_title(cols[i])
plt.tight_layout()
plt.show()
```

```python
def plot_sub(df, cols, ncols=3):
    nrows = np.ceil(len(cols)/ncols).astype(int)
    fig, ax = plt.subplots(figsize=(nrows*3, ncols*2), nrows=nrows, ncols=ncols)
    for i in range(nrows*ncols): 
        if i<len(cols): 
            ax.flat[i].plot(range(len(df)), df[cols[i]].values)
            ax.flat[i].set_title(cols[i])
    plt.tight_layout()
    plt.show()
```

```python
plot_sub(df, [i for i in df.columns if "percent" in i])
```
